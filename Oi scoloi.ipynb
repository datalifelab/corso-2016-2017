<p> RAPPRESENTAZIONE SU PIANO CARTESIA DEI DATI TABELLARI PER ANNO 2014 </p>

```python
#visaulizzazione base (scatterplot) di pandas + annotazioni

fig, ax = plt.subplots()

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = titolo_di_studio_su_elezione_donne


asse_x = dataset.columns[0]
asse_y = dataset.columns[1]
### ----------------------------------------------------------------

dataset.plot(asse_x, asse_y, kind='scatter', ax=ax, figsize = [15,10])

for k, v in dataset[[dataset.columns[0], dataset.columns[1]]] .iterrows():
    ax.annotate(k, v)
    
fig.canvas.draw()
```

```python
#grafico di seaborn (scatterplot) di pandas + annotazioni
#se non fosse installata la libreria seaborn
# da terminale:    pip install seaborn 

import seaborn as sns


sns.set(style="darkgrid")

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = titolo_di_studio_su_elezione_donne


asse_x = dataset.columns[0]
asse_y = dataset.columns[1]
nome_grafico = "xyz.png"
### ----------------------------------------------------------------


color = sns.color_palette()[2]
sns.jointplot(asse_x, asse_y, data=dataset, 
                   color=color, size=7) #xlim=(0, 60), ylim=(0, 12)

def ann(row):
    ind = row[0]
    r = row[1]
    plt.gca().annotate(ind, xy=(r[asse_x], r[asse_y]), 
            xytext=(2,2) , textcoords ="offset points", )

for row in dataset[[dataset.columns[0], dataset.columns[1]]].iterrows():
    ann(row)    

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20, 15)
plt.savefig("uni__bassapaga.png", dpi=300)
```
