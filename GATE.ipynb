<h2> IPOTESI DI PARTENZA</h2>

<p> La nostra ipotesi vuole rilevare se vi è correlazione tra il tasso delle rapine con i valori dell'occupazione e dell'istruzione, su base regionale e in base alla media nazionale. Il tasso di rapine è influenzato dall'occupazione o dall'istruzione? o da nessuna delle due?
Nella prima correlazione abbiamo confrontato il livello di istruzione con gli occupati per avere un quadro generale della situazione italiana e regionale.</p>

```python
#visaulizzazione base (scatterplot) di pandas + annotazioni
Il 
fig, ax = plt.subplots()

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = istruzione_su_tasso_di_occupazione
asse_x = "25-64 fascia di età con diploma"
asse_y = "20-64 fascia di età di occupati"
### ----------------------------------------------------------------

dataset.plot(asse_x, asse_y, kind='scatter', ax=ax, figsize = [15,10])

for k, v in dataset[["25-64 fascia di età con diploma", "20-64 fascia di età di occupati"]].iterrows():
    ax.annotate(k, v)
    
fig.canvas.draw()
```

```python
#grafico di seaborn (scatterplot) di pandas + annotazioni
#se non fosse installata la libreria seaborn
# da terminale:    pip install seaborn 

import seaborn as sns


sns.set(style="darkgrid")

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = istruzione_su_tasso_di_occupazione
asse_x = "25-64 fascia di età con diploma"
asse_y = "20-64 fascia di età di occupati"
nome_grafico = "uni__diplomasuoccupati.png"
### ----------------------------------------------------------------


color = sns.color_palette()[2]
sns.jointplot(asse_x, asse_y, data=dataset, kind="reg",
                   color=color, size=7) #xlim=(0, 60), ylim=(0, 12)

def ann(row):
    ind = row[0]
    r = row[1]
    plt.gca().annotate(ind, xy=(r[asse_x], r[asse_y]), 
            xytext=(2,2) , textcoords ="offset points", )

for row in dataset.iterrows():
    ann(row)    

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20, 15)
plt.savefig("uni__diplomasuoccupati.png", dpi=300)
```

<p> come si evince dal grafico, possiamo trovare una certa relazione tra i due indicatori. le regioni del nord che presentano un alto tasso di diplomati, superiore alla media italiana, hanno un numero di occupazione maggiore rispetto alle regioni del sud, che invece presentano un tasso di diplomati minore rispetto alla media. In entrambe i casi possiamo notare alcune eccezioni. Ad esempio, tra le regioni del nord, la Valle D'Aosta si differenzia perchè a parità di occupati presenta un tasso di diplomati inferiore. Allo stesso modo, il Lazio considerata regione della zonza del centro-sud, presenta valori diverse da quest'ultima, più in linea con le regioni del nord, con addirittura una percentuale di diplomati maggiore.
</p>
