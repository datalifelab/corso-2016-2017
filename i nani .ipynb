```python
#seleziono e concateno gli anni più recenti
occupazione_potere_acquisto = pd.concat([occupazione[2014], capacità_spesa[2014]], axis = 1)
occupazione_potere_acquisto
```

```python
#aggiungo i nomi delle colonne
occupazione_potere_acquisto.columns = ["Occupazione", "Capacità di spesa per famiglia - 2014",]
occupazione_potere_acquisto
```

```python
#Analisi bivariata - correlazione fra le due variabili
occupazione_potere_acquisto
```

```python
#visaulizzazione base (scatterplot) di pandas + annotazioni

fig, ax = plt.subplots()

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = occupazione_potere_acquisto
asse_x = "Occupazione"
asse_y = "Capacità di spesa per famiglia - 2014"
### ----------------------------------------------------------------

dataset.plot(asse_x, asse_y, kind='scatter', ax=ax, figsize = [15,10])

for k, v in dataset.iterrows():
    ax.annotate(k, v)
    
fig.canvas.draw()
```

<p> "dalla tabella è emersa una correlazzione positiva tra la variabile occupazione e la variabile capacità di spesa per famiglia. Nello specifico notiamo le due situazioni limite,sia in positivo che in negativo: Calabraia e Bolzano."</p>

```python
#grafico di seaborn (scatterplot) di pandas + annotazioni
#se non fosse installata la libreria seaborn
# da terminale:    pip install seaborn 

import seaborn as sns


sns.set(style="darkgrid")

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = occupazione_potere_acquisto
asse_x = "Occupazione"
asse_y = "Capacità di spesa per famiglia - 2014"
nome_grafico = "occ__capspes.png"
### ----------------------------------------------------------------


color = sns.color_palette()[2]
sns.jointplot(asse_x, asse_y, data=dataset, kind="reg",
                   color=color, size=7) #xlim=(0, 60), ylim=(0, 12)

def ann(row):
    ind = row[0]
    r = row[1]
    plt.gca().annotate(ind, xy=(r[asse_x], r[asse_y]), 
            xytext=(2,2) , textcoords ="offset points", )

for row in dataset.iterrows():
    ann(row)    

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20, 15)
plt.savefig("occ__capspes.png", dpi=300)
```

<p> "Questo grafico conferma quanto emerso dal grafico precedente, mostrando in più il trend generale." </p>
