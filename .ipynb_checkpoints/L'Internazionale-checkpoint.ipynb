```python
# ---------------------------------------------------------------
# python importa la libreria OS che controlla il sistema operativo
import os 
# ---------------------------------------------------------------
# python importa la libreria Pandas che permette l'importazione ed analisi dei dati: potrà essere richiamata come PD
import pandas as pd 

#importo le librerie di visualizzazione dei grafici
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#importa la libreria per la pulitura del dataset
from corso_lib import restituisci_dataset
```

```python
'''
ZERO DATASET
Tavola 6.1 segue - Incidenza di occupati sovraistruiti per sesso, regione e ripartizione geografica - Anni 2004-2015
'''

totale_sovraistruiti = restituisci_dataset(
    posizione_del_dataset = os.path.join(os.getcwd(), "Appendice-statistica-2016", "03.Lavoro", "Cap_03_Lavoro 06.xlsx"), 
    nome_del_foglio = "REG totale", #inserire il nome del foglio
    riga_intestazione_colonna = 4, # inserire la riga in cui è presente l'intestazione di colonna 
    riga_sottotabella= None, # None - da divisione fra le due tabelle "maschi - femmine" dovrebbe cadere intorno alla riga 36
    tabella = None # None - qui si può mettere "maschi" - "femmine"
)
totale_sovraistruiti
```

```python
'''
ZERO.uno DATASET
REG totale: Tavola 1.1 segue - Tasso di occupazione della popolazione in età 20-64 anni per sesso, regione e ripartizione geografica - Anni 2004-2015 (valori percentuali)

'''

totale_occupati = restituisci_dataset(
    posizione_del_dataset = os.path.join(os.getcwd(), "Appendice-statistica-2016", "03.Lavoro", "Cap_03_Lavoro 01.xlsx"), 
    nome_del_foglio = "REG totale", #inserire il nome del foglio
    riga_intestazione_colonna = 4, # inserire la riga in cui è presente l'intestazione di colonna 
    riga_sottotabella= None, # None - da divisione fra le due tabelle "maschi - femmine" dovrebbe cadere intorno alla riga 36
    tabella = None # None - qui si può mettere "maschi" - "femmine"
)
totale_occupati
```

```python
'''
ZERO.due DATASET
REG totale: Tavola 3.1 segue - Persone di 30-34 anni che hanno conseguito un titolo universitario per sesso, regione e ripartizione geografica - Anni 2004-2015 (valori percentuali)

'''

totale_laureati = restituisci_dataset(
    posizione_del_dataset = os.path.join(os.getcwd(), "Appendice-statistica-2016", "02.Istruzione", "Cap_02_Istruzione 03.xlsx"),
    nome_del_foglio = "REG totale", #inserire il nome del foglio
    riga_intestazione_colonna = 4, # inserire la riga in cui è presente l'intestazione di colonna 
    riga_sottotabella= None, # None - da divisione fra le due tabelle "maschi - femmine" dovrebbe cadere intorno alla riga 36
    tabella = None # None - qui si può mettere "maschi" - "femmine"
)
totale_laureati
```

```python
#seleziono e concateno gli anni più recenti
analisi_sovraistruiti = pd.concat([totale_sovraistruiti[2015], totale_occupati[2015], totale_laureati[2015]], axis = 1)
analisi_sovraistruiti
```

```python
#aggiungo i nomi delle colonne
analisi_sovraistruiti.columns = ["Totale occupati sovraistruiti - 2015", "Tasso occupazione 20-64 - 2015", "Totale Laureati 30-34 - 2015"]
analisi_sovraistruiti
```

```python
#Correlazione multivariata - correlazione fra le due variabili
analisi_sovraistruiti.corr()
```

<h1> ISTRUZIONE E MERCATO DEL LAVORO </h1>
<p> Abbiamo preso in considerazione tre variabili: tasso occupazione 20-64, il tasso dei laureati 30-34 e il tasso degli occupati sovraistruiti tutte relative all anno 2015. L'obiettivo è valutare, in termini di incontro di domanda e offerta di lavoro, se avere una laurea garantisca un'occupazione commisurata allo stesso titolo di studio. </p>


<h1> Scatter plott totale occupati sovraistruiti, totale laureati </h1> 
<p> Un primo dato che emerge dal grafico è un elevto e generalizzato tasso di sovraistruiti in tutte le regioni, compreso tra il 19% del Trentino e il 30% dell'Abruzzo, sia tra le regioni che hanno un alto tasso di laureati che tra quelle in cui si registra un basso tasso di laureati. L unica eccezione è rappresentata dalla provincia autonoma di Bolzano che ha un tasso di occupati sovraistruiti inferiore al 10%.
Guardando al tasso di laureati si rileva un grande divario tra il Centro e il Nord, per i quali si rilevano il 30% e il 28%, rispetto al Mezzogiorno che non arriva al 20%. </p>

```python
#visaulizzazione base (scatterplot) di pandas + annotazioni

fig, ax = plt.subplots()

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = analisi_sovraistruiti
asse_x = "Totale Laureati 30-34 - 2015"
asse_y = "Totale occupati sovraistruiti - 2015"
### ----------------------------------------------------------------

dataset.plot(asse_x, asse_y, kind='scatter', ax=ax, figsize = [15,10])

for k, v in dataset[["Totale Laureati 30-34 - 2015","Totale occupati sovraistruiti - 2015"]] .iterrows():
    ax.annotate(k, v)
    
fig.canvas.draw()
```

<h1> TASSO OCCUPAZIONE, TASSO OCCUPATI SOVRAISTRUITI </h1>
<p> La provincia autonoma di Bolzano conferma anche in questo caso la migliore performance, abbinando al basso tassi di sovraistruiti il tasso di occupazione più elevato tra tutte le regioni, vicino all'80%. In generale tutte le regioni del centro e del nord hanno tassi di occupazione compresi tra il 65% e il 75%, mentre il Mezzogiorno non arriva neanche al 50%. </p>

```python
#visaulizzazione base (scatterplot) di pandas + annotazioni

fig, ax = plt.subplots()

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = analisi_sovraistruiti
asse_x = "Tasso occupazione 20-64 - 2015"
asse_y = "Totale occupati sovraistruiti - 2015"
### ----------------------------------------------------------------

dataset.plot(asse_x, asse_y, kind='scatter', ax=ax, figsize = [15,10])

for k, v in dataset[["Tasso occupazione 20-64 - 2015","Totale occupati sovraistruiti - 2015"]] .iterrows():
    ax.annotate(k, v)
    
fig.canvas.draw()
```

```python
#grafico di seaborn (scatterplot) di pandas + annotazioni
#se non fosse installata la libreria seaborn
# da terminale:    pip install seaborn 

import seaborn as sns


sns.set(style="darkgrid")

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = analisi_sovraistruiti
asse_x = "Tasso occupazione 20-64 - 2015"
asse_y = "Totale occupati sovraistruiti - 2015"
nome_grafico = "uni__sovraistruiti.png"
### ----------------------------------------------------------------


color = sns.color_palette()[2]
sns.jointplot(asse_x, asse_y, data=dataset, kind="reg",
                   color=color, size=7) #xlim=(0, 60), ylim=(0, 12)

def ann(row):
    ind = row[0]
    r = row[1]
    plt.gca().annotate(ind, xy=(r[asse_x], r[asse_y]), 
            xytext=(2,2) , textcoords ="offset points", )

for row in dataset.iterrows():
    ann(row)    

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20, 15)
plt.savefig("uni__bassapaga.png", dpi=300)
```

<h1> ANALISI REGRESSIONE TASSO OCCUPAZIONE E TOTALE OCCUPATI SOVRAISTRUITI </h1>
<p> Il coefficiente di correlazione tra le due variabili è pari a - 0,105. Il segno meno ci indica che tra tasso di occupazione e tasso di sovraistruiti vi è discordanza. In particolare al diminuite del tasso di occupazione aumenta il tasso di occupati sovraistruiti, anche se la pendenza della retta ci indica che siamo in presenza di una relazione debole. Si può affermare che a bassi tassi di occupazione corrispondano maggiori probabilità di trovare un'occupazione caratterizzata da una qualifica inferiore rispetto al proprio titolo di studio. </p>

<h1> CONCLUSIONI </h1>
<p> Il mercato del lavoro italiano (tasso di occupazione) e i livelli di istruzione ( tasso laureati) presentano marcate differenziazione territoriali tra il Centro-Nord e il Mezzogiorno. Inoltre, con l'unica eccezione della provincia autonoma di Bolzano, si rileva su tutto il territorio nazionale un rilevante missmatch orizzontale: tutte le regioni italiane hanno un elevato tasso di occupati sovraistruiti, compreso tra il 20% e il 30%. Le regioni del Centro nordo presentano alti tassi di occupazioni, alti tassi di laureati mentre quelle del Sud presentano anche livelli occupazionali e tassi di laureati preoccupantemente più bassi.
Osservando le serie storiche, partendo dal 2007, anno di inizio della crisi economica, si rileva che la crisi ha accentuato il divario territoriale sia in termini occupazionali che  in termini di livelli istruzione.
In conclusione è possibile affermare che l'occupazione è spesso caratterizzata da una qualifica più bassa rispetto al livello di istruzione. </p>
