```python
obesit = restituisci_dataset(
    posizione_del_dataset = os.path.join(os.getcwd(), "Appendice-statistica-2016", "08.Benessere soggettivo", "Cap_08_Benessere soggettivo 01.xlsx"),
    nome_del_foglio = "REG totale",
    riga_intestazione_colonna = 4,
    riga_sottotabella= None, # None - da divisione fra le due tabelle "maschi - femmine" dovrebbe cadere intorno alla riga 36
    tabella = None # None - qui si può mettere "maschi" - "femmine"
)
obesit
```

```python
#seleziono e concateno gli anni più recenti
obesità_per_regioni_su_servizi = pd.concat([obesit[2015], servizi["2013-2014"]], axis = 1)
obesità_per_regioni_su_servizi
```

```python
obesità_per_regioni_su_servizi["2013-2014"] = obesità_per_regioni_su_servizi["2013-2014"].apply(lambda x: int(x))
```

```python
#aggiungo i nomi delle colonne
obesità_per_regioni_su_servizi.columns = [ "Proporzione di persone di 18 anni in su in sovrappeso o obese", "mobilità giorni feriali"]
obesità_per_regioni_su_servizi
```

```python
#visaulizzazione base (scatterplot) di pandas + annotazioni

fig, ax = plt.subplots()

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = obesità_per_regioni_su_servizi


#"mobilità giorni feriali", "Proporzione di persone di 18 anni in su in sovrappeso o obese"

asse_x = "mobilità giorni feriali"
asse_y = "Proporzione di persone di 18 anni in su in sovrappeso o obese"
### ----------------------------------------------------------------

dataset.plot(asse_x, asse_y, kind='scatter', ax=ax, figsize = [15,10])

for k, v in dataset.iterrows():
    ax.annotate(k, v)
    
fig.canvas.draw()
```

<h1> correlazione tra obesità e mobilità </h1>

    <p> Dal grafico emerge che le regioni del Nord Italia (in particolare Trentino Alto Adige, Friuli venezia Giulia e Valle d'aosta) hanno un tasso di obesità maggiore rispetto alle regioni del Sud, nello specifico le tre con la minore obesità sono Campania, Basilicata e Molise. A fronte di questi dati però non si riscontra una correlazione tra un'alta mobilità nei giorni feriali e la scarsa obesità; infatti regioni come Molise e Basilicata, nonostante presentino una scarsa obesità, hanno un livello di mobilità inferiore rispetto a regioni del Nord, come Lombardia e Liguria. Emerge quindi  una correlazione tra la maggiore mobilità e il minor tasso di obesità, per quanto ci siano delle contraddizioni tra aree geografiche. Quello che si nota dal grafico in conclusione è che un minor tasso di mobilità comporta necessariamente una maggiore obesità; allo stesso tempo però un basso tasso di obesità non è obbligatoriamente determinato da un'alta mobilità quindi la mobilità incide sull'obesità ma non è l'unico fattore.  </p>

```python
#grafico di seaborn (scatterplot) di pandas + annotazioni
#se non fosse installata la libreria seaborn
# da terminale:    pip install seaborn 

import seaborn as sns


sns.set(style="darkgrid")

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = obesità_per_regioni_su_servizi

asse_x = "mobilità giorni feriali"
asse_y = "Proporzione di persone di 18 anni in su in sovrappeso o obese"
nome_grafico = "obesità_servizi.png"
### ----------------------------------------------------------------


color = sns.color_palette()[2]
sns.jointplot(asse_x, asse_y, data=dataset, kind="reg",
                   color=color, size=7) #xlim=(0, 60), ylim=(0, 12)

def ann(row):
    ind = row[0]
    r = row[1]
    plt.gca().annotate(ind, xy=(r[asse_x], r[asse_y]), 
            xytext=(2,2) , textcoords ="offset points", )

for row in dataset.iterrows():
    ann(row)    

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20, 15)
plt.savefig("obesità_servizi.png", dpi=300)
```
